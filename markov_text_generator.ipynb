```python
import random
from collections import defaultdict
```

```python
text = """
data science is fun
data analysis is important
machine learning is part of data science
data science uses statistics
"""
```

```python
text = """
data science is an interdisciplinary field
data science uses statistics and programming
data science helps in decision making
data analysis is an important part of data science
data analysis focuses on finding patterns
machine learning is a subset of data science
machine learning uses algorithms to learn from data
machine learning improves predictions over time
artificial intelligence uses machine learning
data engineering prepares data for analysis
data pipelines move data between systems
data quality is critical for reliable analytics
analytics helps organizations understand performance
analytics supports business decisions
statistics provides mathematical foundations for data science
programming is essential for data analysis
python is widely used in data science
power bi is used for data visualization
dashboards help stakeholders track metrics
data visualization simplifies complex data
"""
```

```python
words = text.lower().split()
```

```python
markov_chain = defaultdict(list)

for i in range(len(words) - 1):
    current_word = words[i]
    next_word = words[i + 1]
    markov_chain[current_word].append(next_word)
```

```python
def generate_text(chain, start_word, length=20):
    word = start_word
    generated_words = [word]

    for _ in range(length - 1):
        if word not in chain:
            break
        word = random.choice(chain[word])
        generated_words.append(word)

    return " ".join(generated_words)
```

```python
output = generate_text(markov_chain, start_word="data", length=15)
print(output)
```
